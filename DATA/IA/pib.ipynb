```python
import pandas as pd


medals_data = {
    'country_name': ['USA', 'China', 'Russia', 'Germany', 'France', 'Japan', 'Australia', 'Canada', 'Italy', 'Brazil'],
    'gold': [39, 38, 20, 10, 10, 27, 17, 4, 10, 7],
    'silver': [41, 32, 28, 11, 12, 14, 7, 6, 10, 6],
    'bronze': [33, 18, 23, 16, 11, 17, 22, 14, 20, 8],
    'total_medals': [113, 88, 71, 37, 33, 58, 46, 24, 40, 21]
}

medals_df = pd.DataFrame(medals_data)
```

```python

economic_data = {
    'country_name': ['USA', 'China', 'Russia', 'Germany', 'France', 'Japan', 'Australia', 'Canada', 'Italy', 'Brazil'],
    'GDP': [21427700, 14342903, 1663933, 3845630, 2715518, 5081770, 1392687, 1647124, 2003093, 1444733],  # en millions de dollars
    'population': [331002651, 1439323776, 145912025, 83783942, 65273511, 126476461, 25499884, 37742154, 60244639, 212559417]  # population en 2020
}

economic_df = pd.DataFrame(economic_data)
```

```python
combined_df = pd.merge(medals_df, economic_df, on='country_name', how='inner')

print(combined_df)
```

```python

columns_of_interest = ['gold', 'silver', 'bronze', 'total_medals', 'GDP', 'population']


correlation_matrix = combined_df[columns_of_interest].corr()


print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap des Corrélations entre Performances Olympiques et Variables Économiques')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=combined_df, x='GDP', y='total_medals', hue='country_name', palette='viridis', s=100)
plt.title('Corrélation entre le PIB et le Nombre Total de Médailles')
plt.xlabel('PIB (en millions de dollars)')
plt.ylabel('Nombre Total de Médailles')
plt
```
