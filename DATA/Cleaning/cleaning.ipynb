```python
df_athletes = pd.read_json("olympic_athletes.json")
df_athletes
```

```python
df_athletes["athlete_medals"] = df_athletes['athlete_medals'].str.replace("\n", "")
```

```python
def extract_gold(palmares):
    if palmares != None:
        match = re.search(r'(\d+)G', palmares)
        return int(match.group(1)) if match else 0
    else:
        return 0

def extract_silver(palmares):
    if palmares != None:
        match = re.search(r'(\d+)S', palmares)
        return int(match.group(1)) if match else 0
    else:
        return 0

def extract_bronze(palmares):
    if palmares != None:
        match = re.search(r'(\d+)B', palmares)
        return int(match.group(1)) if match else 0
    else:
        return 0

df_athletes['gold_medals'] = df_athletes['athlete_medals'].apply(extract_gold)
df_athletes['silver_medals'] = df_athletes['athlete_medals'].apply(extract_silver)
df_athletes["bronze_medals"] = df_athletes['athlete_medals'].apply(extract_bronze)
```
